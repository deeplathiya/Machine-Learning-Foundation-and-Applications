```python
# Load the training set and apply augmentations
train_set = datasets.CIFAR10(root='/', train=True, transform=transforms.Compose([
    transforms.ToTensor(),
    ]), download=True)
# Load the validation set
val_set = datasets.CIFAR10(root='/', train=False, transform=transforms.Compose([
    transforms.ToTensor(),
    ]))
```
